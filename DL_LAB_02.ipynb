```python
import numpy as np
x = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])
epochs = 10
weight = np.zeros(2)
bias = 0
for epoch in range(epochs):
    error_count = 0
    print(f"\nEpoch {epoch+1}")
    for i, j in zip(x, y):
        z = np.dot(i, weight) + bias
        y_pred = 1 if z >= 0 else 0
        error = j - y_pred
        print("error:", error)
        if error != 0:
            weight = weight + error * i
            bias = bias + error
            error_count += 1
        print(weight, bias)
    if error_count == 0:
        print("Network converged")
        break
```
